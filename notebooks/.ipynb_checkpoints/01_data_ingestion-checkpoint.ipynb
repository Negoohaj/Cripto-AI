## 1. Importar Bibliotecas

```python
# Importar bibliotecas necessÃ¡rias
import yfinance as yf
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import os

# ConfiguraÃ§Ãµes
import warnings
warnings.filterwarnings('ignore')

print("âœ… Bibliotecas importadas com sucesso!")
```

## 2. Configurar ParÃ¢metros de Coleta

```python
# Definir criptomoedas para anÃ¡lise
CRYPTO_SYMBOLS = ['BTC-USD', 'ETH-USD']

# Definir perÃ­odo de coleta (Ãºltimos 5 anos)
END_DATE = datetime.now()
START_DATE = END_DATE - timedelta(days=5*365)  # 5 anos

# Intervalo de dados: '1d' (diÃ¡rio), '1h' (horÃ¡rio), '1m' (minuto)
INTERVAL = '1d'

# DiretÃ³rio para salvar dados
DATA_DIR = '../data/raw'
os.makedirs(DATA_DIR, exist_ok=True)

print(f"ğŸ“… PerÃ­odo: {START_DATE.strftime('%Y-%m-%d')} atÃ© {END_DATE.strftime('%Y-%m-%d')}")
print(f"ğŸ“Š Criptomoedas: {', '.join(CRYPTO_SYMBOLS)}")
print(f"â±ï¸  Intervalo: {INTERVAL}")
```

## 3. Coletar Dados do Yahoo Finance

```python
def fetch_crypto_data(symbol, start_date, end_date, interval='1d'):
    """
    Coleta dados histÃ³ricos de uma criptomoeda usando yfinance.
    
    Args:
        symbol (str): SÃ­mbolo da criptomoeda (ex: 'BTC-USD')
        start_date (datetime): Data inicial
        end_date (datetime): Data final
        interval (str): Intervalo dos dados ('1d', '1h', '1m')
    
    Returns:
        pd.DataFrame: DataFrame com os dados histÃ³ricos
    """
    print(f"\nğŸ” Coletando dados de {symbol}...")
    
    try:
        # Baixar dados do Yahoo Finance
        ticker = yf.Ticker(symbol)
        df = ticker.history(
            start=start_date,
            end=end_date,
            interval=interval
        )
        
        # Adicionar coluna com o sÃ­mbolo
        df['Symbol'] = symbol
        
        # Reset index para tornar Date uma coluna
        df = df.reset_index()
        
        print(f"âœ… {len(df)} registros coletados para {symbol}")
        print(f"ğŸ“Š PerÃ­odo: {df['Date'].min()} atÃ© {df['Date'].max()}")
        
        return df
    
    except Exception as e:
        print(f"âŒ Erro ao coletar dados de {symbol}: {str(e)}")
        return None


# Coletar dados para todas as criptomoedas
crypto_data = {}

for symbol in CRYPTO_SYMBOLS:
    df = fetch_crypto_data(symbol, START_DATE, END_DATE, INTERVAL)
    if df is not None:
        crypto_data[symbol] = df

print(f"\nâœ… Coleta finalizada! {len(crypto_data)} criptomoedas coletadas.")
```

## 4. Explorar Dados Coletados

```python
# Visualizar primeiras linhas do Bitcoin
if 'BTC-USD' in crypto_data:
    print("\nğŸ“Š Primeiras linhas do Bitcoin (BTC-USD):")
    display(crypto_data['BTC-USD'].head(10))
    
    print("\nğŸ“ˆ InformaÃ§Ãµes do DataFrame:")
    crypto_data['BTC-USD'].info()
    
    print("\nğŸ“Š EstatÃ­sticas descritivas:")
    display(crypto_data['BTC-USD'].describe())
```

```python
# Visualizar primeiras linhas do Ethereum
if 'ETH-USD' in crypto_data:
    print("\nğŸ“Š Primeiras linhas do Ethereum (ETH-USD):")
    display(crypto_data['ETH-USD'].head(10))
    
    print("\nğŸ“ˆ InformaÃ§Ãµes do DataFrame:")
    crypto_data['ETH-USD'].info()
    
    print("\nğŸ“Š EstatÃ­sticas descritivas:")
    display(crypto_data['ETH-USD'].describe())
```

## 5. Verificar Qualidade dos Dados

```python
# Verificar valores nulos
for symbol, df in crypto_data.items():
    print(f"\nğŸ” Verificando valores nulos em {symbol}:")
    null_counts = df.isnull().sum()
    print(null_counts[null_counts > 0] if null_counts.sum() > 0 else "âœ… Nenhum valor nulo encontrado!")
    
    # Verificar duplicatas
    duplicates = df.duplicated().sum()
    print(f"\nğŸ“‹ Duplicatas: {duplicates}")
```

## 6. Salvar Dados

```python
# Salvar dados em formato CSV e Parquet
for symbol, df in crypto_data.items():
    # Nome do arquivo limpo
    clean_symbol = symbol.replace('-', '_').lower()
    
    # Salvar como CSV
    csv_path = f"{DATA_DIR}/{clean_symbol}_historical.csv"
    df.to_csv(csv_path, index=False)
    print(f"ğŸ’¾ Salvo CSV: {csv_path}")
    
    # Salvar como Parquet (mais eficiente para Big Data)
    parquet_path = f"{DATA_DIR}/{clean_symbol}_historical.parquet"
    df.to_parquet(parquet_path, index=False)
    print(f"ğŸ’¾ Salvo Parquet: {parquet_path}")

print("\nâœ… Todos os dados foram salvos com sucesso!")
```

## 7. Criar Dataset Combinado

```python
# Combinar todos os dados em um Ãºnico DataFrame
if len(crypto_data) > 0:
    combined_df = pd.concat(crypto_data.values(), ignore_index=True)
    
    print(f"\nğŸ“Š Dataset combinado criado com {len(combined_df)} registros")
    print(f"ğŸ’± Criptomoedas: {combined_df['Symbol'].unique()}")
    
    # Salvar dataset combinado
    combined_csv = f"{DATA_DIR}/crypto_combined.csv"
    combined_parquet = f"{DATA_DIR}/crypto_combined.parquet"
    
    combined_df.to_csv(combined_csv, index=False)
    combined_df.to_parquet(combined_parquet, index=False)
    
    print(f"\nğŸ’¾ Dataset combinado salvo:")
    print(f"   - {combined_csv}")
    print(f"   - {combined_parquet}")
    
    display(combined_df.head())
```

## 8. Resumo da Coleta

```python
print("\n" + "="*60)
print("ğŸ“Š RESUMO DA COLETA DE DADOS")
print("="*60)

for symbol, df in crypto_data.items():
    print(f"\nğŸ’° {symbol}:")
    print(f"   ğŸ“… PerÃ­odo: {df['Date'].min()} atÃ© {df['Date'].max()}")
    print(f"   ğŸ“Š Total de registros: {len(df):,}")
    print(f"   ğŸ’µ PreÃ§o mÃ­nimo: ${df['Low'].min():.2f}")
    print(f"   ğŸ’µ PreÃ§o mÃ¡ximo: ${df['High'].max():.2f}")
    print(f"   ğŸ’µ PreÃ§o mÃ©dio: ${df['Close'].mean():.2f}")
    print(f"   ğŸ“ˆ Volume total: {df['Volume'].sum():,.0f}")

print("\n" + "="*60)
print("âœ… Coleta de dados finalizada com sucesso!")
print("ğŸ“ PrÃ³ximo passo: 02_spark_processing.ipynb")
print("="*60)
```
