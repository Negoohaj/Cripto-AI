## 1. Importar Bibliotecas

```python
# Importar bibliotecas necessÃ¡rias
import pandas as pd
import numpy as np
from prophet import Prophet
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta
import warnings
import pickle
import os

warnings.filterwarnings('ignore')

# ConfiguraÃ§Ãµes de visualizaÃ§Ã£o
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

print("âœ… Bibliotecas importadas com sucesso!")
```

## 2. Carregar Dados Processados

```python
# Carregar dados processados do notebook anterior
DATA_PATH = "../data/processed/crypto_processed.csv"
MODEL_DIR = "../models/"
os.makedirs(MODEL_DIR, exist_ok=True)

print("ğŸ“‚ Carregando dados processados...")
df = pd.read_csv(DATA_PATH, parse_dates=['Date'])

print(f"âœ… Carregado: {len(df):,} registros")
print(f"ğŸ’± Criptomoedas: {df['Symbol'].unique()}")
print(f"ğŸ“… PerÃ­odo: {df['Date'].min()} atÃ© {df['Date'].max()}")

df.head()
```

## 3. Preparar Dados para o Prophet

```python
def prepare_prophet_data(df, symbol):
    """
    Prepara dados no formato requerido pelo Prophet.
    Prophet requer colunas 'ds' (data) e 'y' (valor).
    
    Args:
        df (pd.DataFrame): DataFrame com os dados
        symbol (str): SÃ­mbolo da criptomoeda
    
    Returns:
        pd.DataFrame: DataFrame formatado para o Prophet
    """
    # Filtrar dados do sÃ­mbolo especÃ­fico
    df_symbol = df[df['Symbol'] == symbol].copy()
    
    # Selecionar e renomear colunas
    df_prophet = df_symbol[['Date', 'Close']].rename(
        columns={'Date': 'ds', 'Close': 'y'}
    )
    
    # Ordenar por data
    df_prophet = df_prophet.sort_values('ds').reset_index(drop=True)
    
    # Remover valores nulos
    df_prophet = df_prophet.dropna()
    
    return df_prophet


# Preparar dados para Bitcoin
print("\nğŸª™ Preparando dados do Bitcoin...")
df_btc = prepare_prophet_data(df, 'BTC-USD')
print(f"âœ… Bitcoin: {len(df_btc)} registros")

# Preparar dados para Ethereum
print("\nğŸ’ Preparando dados do Ethereum...")
df_eth = prepare_prophet_data(df, 'ETH-USD')
print(f"âœ… Ethereum: {len(df_eth)} registros")

display(df_btc.head())
display(df_btc.tail())
```

## 4. Treinar Modelo Prophet - Bitcoin

```python
print("ğŸ”® Treinando modelo Prophet para Bitcoin...\n")

# Configurar modelo Prophet
model_btc = Prophet(
    daily_seasonality=False,      # Sem sazonalidade diÃ¡ria
    weekly_seasonality=True,      # Com sazonalidade semanal
    yearly_seasonality=True,      # Com sazonalidade anual
    changepoint_prior_scale=0.05, # Flexibilidade da tendÃªncia
    seasonality_prior_scale=10.0, # ForÃ§a da sazonalidade
    interval_width=0.95           # Intervalo de confianÃ§a de 95%
)

# Adicionar sazonalidade mensal customizada
model_btc.add_seasonality(
    name='monthly',
    period=30.5,
    fourier_order=5
)

# Treinar modelo
print("â³ Treinando modelo...")
model_btc.fit(df_btc)
print("âœ… Modelo treinado com sucesso!")

# Salvar modelo
model_btc_path = f"{MODEL_DIR}prophet_btc.pkl"
with open(model_btc_path, 'wb') as f:
    pickle.dump(model_btc, f)
print(f"ğŸ’¾ Modelo salvo: {model_btc_path}")
```

## 5. Fazer PrevisÃµes - Bitcoin

```python
# Criar dataframe para previsÃµes futuras (prÃ³ximos 30 dias)
FORECAST_DAYS = 30

print(f"ğŸ”® Gerando previsÃµes para os prÃ³ximos {FORECAST_DAYS} dias...\n")

# Criar datas futuras
future_btc = model_btc.make_future_dataframe(periods=FORECAST_DAYS, freq='D')
print(f"ğŸ“… PrevisÃµes de {future_btc['ds'].min()} atÃ© {future_btc['ds'].max()}")

# Fazer previsÃµes
forecast_btc = model_btc.predict(future_btc)

# Exibir previsÃµes
print("\nğŸ“Š PrevisÃµes para Bitcoin:")
forecast_columns = ['ds', 'yhat', 'yhat_lower', 'yhat_upper', 'trend']
display(forecast_btc[forecast_columns].tail(10))

# Separar previsÃµes futuras
last_date = df_btc['ds'].max()
future_forecast_btc = forecast_btc[forecast_btc['ds'] > last_date]

print(f"\nğŸ¯ PrevisÃ£o para os prÃ³ximos {FORECAST_DAYS} dias:")
display(future_forecast_btc[forecast_columns])
```

## 6. Visualizar PrevisÃµes - Bitcoin

```python
# Plotar previsÃµes
fig1 = model_btc.plot(forecast_btc, figsize=(15, 6))
plt.title('Bitcoin - PrevisÃ£o de PreÃ§os com Prophet', fontsize=16, fontweight='bold')
plt.xlabel('Data', fontsize=12)
plt.ylabel('PreÃ§o (USD)', fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('../visualizations/btc_prophet_forecast.png', dpi=300, bbox_inches='tight')
plt.show()

print("ğŸ’¾ GrÃ¡fico salvo: ../visualizations/btc_prophet_forecast.png")
```

## 7. Analisar Componentes do Modelo - Bitcoin

```python
# Plotar componentes (tendÃªncia, sazonalidade semanal, anual)
fig2 = model_btc.plot_components(forecast_btc, figsize=(15, 10))
plt.tight_layout()
plt.savefig('../visualizations/btc_prophet_components.png', dpi=300, bbox_inches='tight')
plt.show()

print("ğŸ’¾ GrÃ¡fico salvo: ../visualizations/btc_prophet_components.png")

print("\nğŸ“Š AnÃ¡lise dos Componentes:")
print("   ğŸ“ˆ Trend: TendÃªncia geral de longo prazo")
print("   ğŸ“… Weekly: PadrÃ£o semanal de variaÃ§Ã£o")
print("   ğŸ—“ï¸  Yearly: PadrÃ£o anual de sazonalidade")
print("   ğŸ“† Monthly: PadrÃ£o mensal customizado")
```

## 8. Treinar Modelo Prophet - Ethereum

```python
print("ğŸ”® Treinando modelo Prophet para Ethereum...\n")

# Configurar modelo Prophet
model_eth = Prophet(
    daily_seasonality=False,
    weekly_seasonality=True,
    yearly_seasonality=True,
    changepoint_prior_scale=0.05,
    seasonality_prior_scale=10.0,
    interval_width=0.95
)

# Adicionar sazonalidade mensal
model_eth.add_seasonality(
    name='monthly',
    period=30.5,
    fourier_order=5
)

# Treinar modelo
print("â³ Treinando modelo...")
model_eth.fit(df_eth)
print("âœ… Modelo treinado com sucesso!")

# Salvar modelo
model_eth_path = f"{MODEL_DIR}prophet_eth.pkl"
with open(model_eth_path, 'wb') as f:
    pickle.dump(model_eth, f)
print(f"ğŸ’¾ Modelo salvo: {model_eth_path}")
```

## 9. Fazer PrevisÃµes - Ethereum

```python
print(f"ğŸ”® Gerando previsÃµes para Ethereum...\n")

# Criar datas futuras
future_eth = model_eth.make_future_dataframe(periods=FORECAST_DAYS, freq='D')

# Fazer previsÃµes
forecast_eth = model_eth.predict(future_eth)

# Separar previsÃµes futuras
last_date_eth = df_eth['ds'].max()
future_forecast_eth = forecast_eth[forecast_eth['ds'] > last_date_eth]

print(f"ğŸ¯ PrevisÃ£o para Ethereum nos prÃ³ximos {FORECAST_DAYS} dias:")
display(future_forecast_eth[forecast_columns])
```

```python
# Plotar previsÃµes
fig3 = model_eth.plot(forecast_eth, figsize=(15, 6))
plt.title('Ethereum - PrevisÃ£o de PreÃ§os com Prophet', fontsize=16, fontweight='bold')
plt.xlabel('Data', fontsize=12)
plt.ylabel('PreÃ§o (USD)', fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('../visualizations/eth_prophet_forecast.png', dpi=300, bbox_inches='tight')
plt.show()

print("ğŸ’¾ GrÃ¡fico salvo: ../visualizations/eth_prophet_forecast.png")
```

## 12. Avaliar Performance do Modelo

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

def evaluate_model(df_actual, forecast, name):
    """
    Avalia performance do modelo Prophet.
    """
    # Merge dos dados reais com previsÃµes
    comparison = df_actual.merge(forecast[['ds', 'yhat']], on='ds', how='inner')
    
    # Calcular mÃ©tricas
    mae = mean_absolute_error(comparison['y'], comparison['yhat'])
    rmse = np.sqrt(mean_squared_error(comparison['y'], comparison['yhat']))
    r2 = r2_score(comparison['y'], comparison['yhat'])
    mape = np.mean(np.abs((comparison['y'] - comparison['yhat']) / comparison['y'])) * 100
    
    print(f"\nğŸ“Š MÃ©tricas de AvaliaÃ§Ã£o - {name}:")
    print(f"   MAE (Mean Absolute Error): ${mae:,.2f}")
    print(f"   RMSE (Root Mean Squared Error): ${rmse:,.2f}")
    print(f"   RÂ² Score: {r2:.4f}")
    print(f"   MAPE (Mean Absolute Percentage Error): {mape:.2f}%")
    
    return {'MAE': mae, 'RMSE': rmse, 'R2': r2, 'MAPE': mape}

# Avaliar modelos
metrics_btc = evaluate_model(df_btc, forecast_btc, "Bitcoin")
metrics_eth = evaluate_model(df_eth, forecast_eth, "Ethereum")
```
