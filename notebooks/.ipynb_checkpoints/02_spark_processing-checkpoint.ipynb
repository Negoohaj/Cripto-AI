## 1. Configurar Spark Session

```python
# Importar bibliotecas do PySpark
from pyspark.sql import SparkSession
from pyspark.sql import functions as F
from pyspark.sql.window import Window
from pyspark.sql.types import *
import pandas as pd

# Criar SparkSession (no Databricks, spark jÃ¡ estÃ¡ disponÃ­vel)
# Para execuÃ§Ã£o local, descomente as linhas abaixo:
'''
spark = SparkSession.builder \
    .appName("CryptoAnalysis") \
    .config("spark.sql.extensions", "io.delta.sql.DeltaSparkSessionExtension") \
    .config("spark.sql.catalog.spark_catalog", "org.apache.spark.sql.delta.catalog.DeltaCatalog") \
    .getOrCreate()
'''

print("âœ… Spark Session configurada!")
print(f"ğŸ“Š VersÃ£o do Spark: {spark.version}")
print(f"ğŸ–¥ï¸  Master: {spark.sparkContext.master}")
```

## 2. Carregar Dados no Spark

```python
# Definir caminhos dos dados
# No Databricks, ajuste para o caminho do DBFS: /dbfs/FileStore/...
DATA_PATH = "../data/raw/crypto_combined.parquet"
OUTPUT_PATH = "../data/processed/"

# Carregar dados no Spark
print("ğŸ“‚ Carregando dados no Spark...")
df_spark = spark.read.parquet(DATA_PATH)

# Exibir schema
print("\nğŸ“‹ Schema dos dados:")
df_spark.printSchema()

# Contar registros
total_records = df_spark.count()
print(f"\nğŸ“Š Total de registros: {total_records:,}")

# Mostrar primeiras linhas
print("\nğŸ“Š Primeiras linhas:")
df_spark.show(5)
```

## 3. Limpeza e PreparaÃ§Ã£o dos Dados

```python
# Remover valores nulos
df_clean = df_spark.dropna()

# Converter Date para formato de data correto
df_clean = df_clean.withColumn("Date", F.to_date(F.col("Date")))

# Ordenar por sÃ­mbolo e data
df_clean = df_clean.orderBy("Symbol", "Date")

# Verificar dados limpos
print(f"âœ… Registros apÃ³s limpeza: {df_clean.count():,}")
df_clean.show(5)
```

## 4. Calcular MÃ©dias MÃ³veis (Moving Averages)

```python
# Definir janelas para mÃ©dias mÃ³veis por sÃ­mbolo
# Window particionado por Symbol e ordenado por Date
window_7 = Window.partitionBy("Symbol").orderBy("Date").rowsBetween(-6, 0)
window_30 = Window.partitionBy("Symbol").orderBy("Date").rowsBetween(-29, 0)
window_90 = Window.partitionBy("Symbol").orderBy("Date").rowsBetween(-89, 0)

print("ğŸ“Š Calculando mÃ©dias mÃ³veis...")
print("   - MA7: MÃ©dia mÃ³vel de 7 dias")
print("   - MA30: MÃ©dia mÃ³vel de 30 dias")
print("   - MA90: MÃ©dia mÃ³vel de 90 dias")

# Calcular mÃ©dias mÃ³veis usando Spark Window Functions
df_with_ma = df_clean.withColumn(
    "MA7", F.avg("Close").over(window_7)
).withColumn(
    "MA30", F.avg("Close").over(window_30)
).withColumn(
    "MA90", F.avg("Close").over(window_90)
)

# Calcular volatilidade (desvio padrÃ£o mÃ³vel de 30 dias)
df_with_ma = df_with_ma.withColumn(
    "Volatility_30d", F.stddev("Close").over(window_30)
)

# Calcular retorno diÃ¡rio (variaÃ§Ã£o percentual)
window_prev = Window.partitionBy("Symbol").orderBy("Date")
df_with_ma = df_with_ma.withColumn(
    "Daily_Return",
    ((F.col("Close") - F.lag("Close", 1).over(window_prev)) / F.lag("Close", 1).over(window_prev)) * 100
)

print("\nâœ… MÃ©dias mÃ³veis calculadas!")
df_with_ma.select("Symbol", "Date", "Close", "MA7", "MA30", "MA90", "Volatility_30d", "Daily_Return").show(10)
```

## 5. AnÃ¡lises Agregadas

```python
# EstatÃ­sticas por criptomoeda
print("ğŸ“Š EstatÃ­sticas agregadas por criptomoeda:\n")

stats = df_with_ma.groupBy("Symbol").agg(
    F.count("*").alias("Total_Days"),
    F.min("Date").alias("Start_Date"),
    F.max("Date").alias("End_Date"),
    F.min("Low").alias("Min_Price"),
    F.max("High").alias("Max_Price"),
    F.avg("Close").alias("Avg_Price"),
    F.sum("Volume").alias("Total_Volume"),
    F.avg("Daily_Return").alias("Avg_Daily_Return"),
    F.stddev("Daily_Return").alias("Volatility_Daily_Return")
).orderBy("Symbol")

stats.show(truncate=False)
```

## 6. AnÃ¡lise Mensal

```python
# Adicionar colunas de ano e mÃªs
df_monthly = df_with_ma.withColumn("Year", F.year("Date")) \
                       .withColumn("Month", F.month("Date"))

# AgregaÃ§Ãµes mensais
monthly_stats = df_monthly.groupBy("Symbol", "Year", "Month").agg(
    F.first("Close").alias("Open_Price"),
    F.last("Close").alias("Close_Price"),
    F.min("Low").alias("Low_Price"),
    F.max("High").alias("High_Price"),
    F.avg("Close").alias("Avg_Price"),
    F.sum("Volume").alias("Total_Volume")
).orderBy("Symbol", "Year", "Month")

# Calcular retorno mensal
monthly_stats = monthly_stats.withColumn(
    "Monthly_Return",
    ((F.col("Close_Price") - F.col("Open_Price")) / F.col("Open_Price")) * 100
)

print("ğŸ“… AnÃ¡lise mensal:")
monthly_stats.show(20)
```

## 7. Identificar TendÃªncias

```python
# Identificar sinais de compra/venda baseados em cruzamento de mÃ©dias mÃ³veis
# Sinal de compra: MA7 cruza MA30 para cima
# Sinal de venda: MA7 cruza MA30 para baixo

df_signals = df_with_ma.withColumn(
    "Signal",
    F.when(
        (F.col("MA7") > F.col("MA30")) & 
        (F.lag("MA7", 1).over(window_prev) <= F.lag("MA30", 1).over(window_prev)),
        "BUY"
    ).when(
        (F.col("MA7") < F.col("MA30")) & 
        (F.lag("MA7", 1).over(window_prev) >= F.lag("MA30", 1).over(window_prev)),
        "SELL"
    ).otherwise("HOLD")
)

# Filtrar apenas sinais de compra e venda
trading_signals = df_signals.filter(F.col("Signal").isin(["BUY", "SELL"]))

print("ğŸ”” Sinais de Trading identificados:")
trading_signals.select("Symbol", "Date", "Close", "MA7", "MA30", "Signal").show(20)
```

## 8. Salvar Dados Processados

```python
# Salvar dados com mÃ©dias mÃ³veis
print("ğŸ’¾ Salvando dados processados...")

# Salvar como Parquet
output_parquet = f"{OUTPUT_PATH}crypto_with_indicators.parquet"
df_with_ma.write.mode("overwrite").parquet(output_parquet)
print(f"âœ… Salvo: {output_parquet}")

# Salvar agregaÃ§Ãµes mensais
monthly_output = f"{OUTPUT_PATH}monthly_aggregations.parquet"
monthly_stats.write.mode("overwrite").parquet(monthly_output)
print(f"âœ… Salvo: {monthly_output}")

# Salvar sinais de trading
signals_output = f"{OUTPUT_PATH}trading_signals.parquet"
trading_signals.write.mode("overwrite").parquet(signals_output)
print(f"âœ… Salvo: {signals_output}")

# Opcional: Salvar como Delta Lake (recomendado para Databricks)
# df_with_ma.write.format("delta").mode("overwrite").save("/delta/crypto_indicators")

print("\nâœ… Processamento finalizado!")
```

## 9. Converter para Pandas (para anÃ¡lises posteriores)

```python
# Converter para Pandas para usar nos prÃ³ximos notebooks
print("ğŸ”„ Convertendo Spark DataFrame para Pandas...")

# Selecionar apenas dados recentes para evitar problemas de memÃ³ria
df_recent = df_with_ma.filter(F.col("Date") >= F.date_sub(F.current_date(), 730))  # 2 anos

df_pandas = df_recent.toPandas()

# Salvar como CSV para prÃ³ximos notebooks
csv_output = f"{OUTPUT_PATH}crypto_processed.csv"
df_pandas.to_csv(csv_output, index=False)
print(f"âœ… Salvo CSV: {csv_output}")

print(f"\nğŸ“Š Dataset Pandas: {len(df_pandas):,} registros")
df_pandas.head()
```
